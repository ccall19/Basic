### 权限详解

-rw-r--r--
│ │││ │││
│ ││└─┘││└─ 其他用户（others）权限
│ │└──┘└─── 所属组（group）权限
│ └────────── 所有者（owner）权限
└──────────── 文件类型（-=普通文件，d=目录）

# 关于 chmod

### 基础语法

```python
chmod [选项] 权限 文件
```

## chmod 权限字符说明

### 用户类型

| 字符 | 含义 | 缩写 |
|------|------|------|
| `u` | 用户（user/所有者） | User |
| `g` | 组（group） | Group |
| `o` | 其他（others） | Others |
| `a` | 所有（all） | All |

### 操作符

| 符号 | 含义 | 说明 |
|------|------|------|
| `+` | 添加权限 | 在现有基础上增加 |
| `-` | 移除权限 | 在现有基础上删除 |
| `=` | 设置权限 | 覆盖现有权限 |

### 权限类型

| 字符 | 权限 | 数值 |
|------|------|------|
| `r` | 读（read） | 4 |
| `w` | 写（write） | 2 |
| `x` | 执行（execute） | 1 |
| `-` | 无权限 | 0 |

### 常用组合

| 命令 | 结果 | 用途 |
|------|------|------|
| `chmod +x file` | `-rwxr-xr-x` | 所有人可执行 |
| `chmod u+x file` | `-rwxrw-r--` | 只有所有者可执行 |
| `chmod 755 file` | `-rwxr-xr-x` | 标准可执行文件 |
| `chmod 644 file` | `-rw-r--r--` | 标准文本文件 |
