# FastAPI Pydantic 模型
参看 [菜鸟文档](https://www.runoob.com/fastapi/fastapi-pydantic.html)
以及 [官方文档](https://docs.pydantic.dev/latest/)

Pydantic 用于数据验证和序列化 。        
使用 Pydantic 定义一个模型非常简单，只需创建一个继承自 pydantic.BaseModel 的类，并在其中定义字段。字段的类型可以是任何有效的 Python 类型，也可以是 Pydantic 内置的类型。

from pydantic import BaseModel

````py
class Item(BaseModel):
    name: str
    description: str = None
    price: float
    tax: float = None
````
