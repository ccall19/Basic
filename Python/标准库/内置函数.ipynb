### next(iterator[, default])
通过调用 iterator 的 __next__() 方法获取下一个元素。如果迭代器耗尽，则返回给定的 default，如果没有默认值则触发 StopIteration。

```python
oriList = [1, 2, 3, 4, 5]
it = iter(oriList)
print(dir(it))
print(next(it))  # 输出 1
print(next(it))  # 输出 2
print('*' * 10)
for item in it:
    print(item)  # 输出 3, 4, 5
```

```python
print(any(name == '__next__' for name in dir(it)))
```
