Python 运行时不强制要求函数与变量类型注解。typing这个模块提供对类型提示的运行时支持。最基本的支持包括 Any , Union , Callable , TypeVar , 和 Generic 。

`````python
def greeting(name: str) -> str:
    return 'Hello ' + name
`````

### typing.Union
联合类型； Union[X, Y] 等价于 X | Y ，意味着满足 X 或 Y 之一。要定义一个联合类型，可以使用类似 Union[int, str] 或简写 int | str。建议使用这种简写。

### typing.Type
Type 本身也可以是一种类型，比如：       
````py
a = 3         # Has type 'int'
b = int       # Has type 'Type[int]'
c = type(a)   # Also has type 'Type[int]'
````

### typing.Literal
给定一组字面量，检查参数值 是否在其中，比如：

````py
def validate_simple(data: Any) -> Literal[True]:  # always returns True
    ...

MODE = Literal['r', 'rb', 'w', 'wb']
def open_helper(file: str, mode: MODE) -> str:
    ...

open_helper('/some/path', 'r')  # Passes type check
open_helper('/other/path', 'typo')  # Error in type checker
````

### typing.TypedDict
细节参考：https://docs.python.org/zh-cn/3.10/library/typing.html?highlight=typing#typing.TypedDict      
把类型提示添加至字典的特殊构造器。在运行时，它是纯 dict。比如：

````py
class Point2D(TypedDict):
    x: int
    y: int
    label: str

a: Point2D = {'x': 1, 'y': 2, 'label': 'good'}  # OK
b: Point2D = {'z': 3, 'label': 'bad'}           # Fails type check

assert Point2D(x=1, y=2, label='first') == dict(x=1, y=2, label='first')
````
