# 3. 命令行参数
通用实用程序脚本通常需要处理命令行参数。这些参数作为列表存储在 sys 模块的 argv 属性中。例如，以下输出来自在命令行运行 python demo.py one two three
````py
>>>
>>> import sys
>>> print(sys.argv)
['demo.py', 'one', 'two', 'three']
````
## 但实际中常用：argparse      
实现主要是三步骤，声明一个解析器argparse.ArgumentParser；添加参数parser.add_argument；解析参数parser.parse_args()       
````py
import argparse

def main():
    parser = argparse.ArgumentParser(description="一个简单的示例脚本")
    parser.add_argument('name', help='输入姓名')
    parser.add_argument('age', type=int, help='输入年龄')
    parser.add_argument('--city', default='未知', help='输入城市 (可选)')
    
    args = parser.parse_args()
    
    print(f"姓名: {args.name}, 年龄: {args.age}, 城市: {args.city}")

if __name__ == '__main__':
    main()
````
