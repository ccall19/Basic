# Schema 定义了 Collections 的数据结构。

Schema 设计的目标：让检索更精准、更高效

关键：不只是存向量，还要存好元数据，支持过滤和溯源

## 创建 一个 Schema

```python
from pymilvus import MilvusClient, DataType

schema = MilvusClient.create_schema()
```

## 添加 主索引 字段

- 每个 Collection 必须有一个主字段。

- 主字段值不能为空。

- 数据类型必须在创建时指定，以后不能更改。

支持 数据类型： `INT64`、VARCHAR

自动创建：auto_id=True, 



```python
schema.add_field(
    field_name="my_id",
    datatype=DataType.INT64,
    is_primary=True,
    auto_id=False,
)
```

## 添加 向量 字段

```python
schema.add_field(
    field_name="my_vector",
    datatype=DataType.FLOAT_VECTOR,
    dim=5
)
```

## 添加标量字段

### 字符串、整数、布尔值

```python
schema.add_field(
    field_name="my_varchar",
    datatype=DataType.VARCHAR,
    max_length=512
)

schema.add_field(
    field_name="my_int64",
    datatype=DataType.INT64,
)

schema.add_field(
    field_name="my_bool",
    datatype=DataType.BOOL,
)
```

## 添加 json 字段

```python
schema.add_field(
    field_name="my_json",
    datatype=DataType.JSON,
)
```

## 添加 数组 字段

```python
### 数组字段中所有元素的数据类型应相同。

schema.add_field(
    field_name="my_array",
    datatype=DataType.ARRAY,
    element_type=DataType.VARCHAR,
    max_capacity=5,
    max_length=512,
)
```
