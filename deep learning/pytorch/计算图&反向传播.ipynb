## 1、基础部分

$$ y = f(x) = 6x^2 + 2x + 4 $$

$$ \frac{dy}{dx} = 12x + 2 $$

```python
x = torch.tensor(3., requires_grad=True)
print(x.data)
print(x.grad)
print(x.requires_grad)
print(x.is_leaf)
```

```python
y = 6 * x**2 + 2 * x + 4
print(y.data)
print(y.grad)
print(y.requires_grad)
print(y.is_leaf)
```

```python
y.backward()
print(x.data)
print(x.grad)
```
