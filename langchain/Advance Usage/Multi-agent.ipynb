# Multi-agent
主要是体现了 解耦 的思想，将多个 更简单、更专注 的 Agent 协调 为 workflow

问题场景：      
❌ 单个 Agent 有太多工具 → 工具选择效果差       
❌ Context 太大 → Agent 无法有效跟踪        
❌ 任务需要专业化 → 需要不同的"专家"        

解决方案：      
✅ 分解为多个小的、专注的 Agent     
✅ 让它们协作解决问题       
✅ 每个 Agent 专攻一个领域      

## 两大模式
### 1. Tool Calling（工具调用） - 中央控制
一个主 Agent（Controller）把其他 Agent 当作工具来调用。

### 2. Handoffs（交接） - 去中央化
代理之间直接转移控制权。活跃的代理会改变，用户可以和不同的代理交互。
