## 1 天气查询Agent
- FastAPI 流式输出：借助 StreamingResponse 返回可迭代对象，框架会以 chunked 方式逐块写入 HTTP 响应，从而不用等整个结果生成完毕。
- 为何选 SSE：Server-Sent Events 在浏览器里有内建 EventSource 客户端，协议简单、基于 HTTP、天然支持自动重连，适合单向从服务端向前端持续推送文本片段。
- 消息与 token 级流：LangChain 的 stream() 会持续产出 token（或消息片段）；后端把每个片段格式化成 SSE（data: ...\n\n），前端收到事件后即时追加到界面，形成“逐字打字”的体验。
- 与 WebSocket 比较：WebSocket 支持双向通信与自定义协议，但需要额外握手、心跳和兼容处理；SSE 只需普通 HTTP，代码更轻，足够覆盖“服务器→客户端”场景。
- 前端处理：使用 EventSource 或手动 fetch+流式 reader 解析 SSE，将分段文本累积后交给 Markdown 渲染器，即可实时显示换行等格式。
