### Stream
调用  stream() 方法 返回一个迭代器。和 invoke() 方法 单纯返回一条  [AIMessage](https://reference.langchain.com/python/langchain/messages/?_gl=1*lsmlor*_gcl_au*MTMyODY3MTI4Ni4xNzYyOTQ1MzMw*_ga*MTQ0ODQ4NzAyNS4xNzYyOTQ1MzMw*_ga_47WX3HKKY2*czE3NjMxOTczNDQkbzE4JGcwJHQxNzYzMTk3MzQ0JGo2MCRsMCRoMA..#langchain.messages.AIMessage) 不同，stream() 返回一个 [AIMessageChunk](https://reference.langchain.com/python/langchain/messages/?_gl=1*lsmlor*_gcl_au*MTMyODY3MTI4Ni4xNzYyOTQ1MzMw*_ga*MTQ0ODQ4NzAyNS4xNzYyOTQ1MzMw*_ga_47WX3HKKY2*czE3NjMxOTczNDQkbzE4JGcwJHQxNzYzMTk3MzQ0JGo2MCRsMCRoMA..#langchain.messages.AIMessageChunk) 对象

### Batch
默认  batch() 返回所有输入最终的输出，如果想要流式输出每一个 Input 的 Output，则使用 batch_as_completed()：
````py
responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
])
for response in responses:
    print(response)

for response in model.batch_as_completed([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is quantum computing?"
]):
    print(response)
````

## Tool calling
在 LangChain 中，工具的调用除了在创建 Agent 时 用 列表指定，还可以为模型 绑定 tools，通过 bind_tools():
````py
model_with_tools = model.bind_tools([get_weather])  
response = model_with_tools.invoke("What's the weather like in Boston?")

model_with_tools = model.bind_tools([tool_1], tool_choice="any") # Force use of any tool
model_with_tools = model.bind_tools([tool_1], tool_choice="tool_1") # Force use of specific tools
````

## 关于多模态
You can pass non-textual data to a model by providing content blocks.
````py
response = model.invoke("Create a picture of a cat")
print(response.content_blocks)
# [
#     {"type": "text", "text": "Here's a picture of a cat"},
#     {"type": "image", "base64": "...", "mime_type": "image/jpeg"},
# ]
````
