LangChain 包 有六大 子包：
- langchain
- langchain-core
- langchain-text-splitters
- langchain-mcp-adapters
- langchain-tests
- langchain-classic

# LangChian
定义交互逻辑
- Agents: Agent interface and factory.
- Middleware: Middleware for agents.
- Models: Chat model interfaces.
- Messages: Chat message types.
- Tools: Tool interface and factory.
- Embeddings: Embedding model interfaces.

# langchain-core
核心运行时逻辑 + 基本接口
- Caches: Caching mechanisms.
- Callbacks: Callback handlers and management.
- Documents: Document abstractions.
- Embeddings: Embedding abstractions.
- Exceptions: Common LangChain exception types.
- Language models: Base interfaces for language models.
