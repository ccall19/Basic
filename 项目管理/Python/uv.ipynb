参看： 项目管理/Python/venv.ipynb 中，一系列纯手动流程：
````sh
python3 -m venv .venv # 创建虚拟环境
source .venv/bin/activate # 启动虚拟环境
edit pyproject.toml # 编辑配置文件
pip install -e . # 安装相关依赖
````

uv 加入一个新的依赖 仅需要 一条命令(包括安装间接依赖)：

````sh
uv add flask
````

使用 uv 依然可以用传统的 方式来运行代码：

````sh
source .venv/bin/activate # 激活虚拟环境
python main.py # python 运行
````

但 uv 也提供了 一个更加简洁的 做法：

````sh
uv run main.py
````

其作用是在虚拟环境的 上下文中执行命令，不需要我们手动的激活环境。   

### 想要查看项目依赖？

````sh
uv tree
````

## uv 虚拟环境与磁盘占用
uv 使用硬链接机制共享全局缓存中的包，硬链接不占额外空间。       

验证硬链接：

````sh
# 检查链接数（>1 表示有硬链接）
ls -li /path/to/file

# 查看实际物理占用
du -sh ~/.cache/uv/
````
