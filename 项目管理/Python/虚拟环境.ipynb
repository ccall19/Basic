### 输出当前虚拟环境的 包 到一个文件中

````sh
pip freeze > requirements.txt
````

### 安装

````sh
pip install -r requirements.txt
````

pip 的 缺点是 无法区分 直接依赖 和 间接依赖，尤其是在卸载的时候，卸载直接依赖并不会把间接依赖也卸载掉。     

因此 ，现代的 管理工具 是 使用 `pyproject.toml` 文件 来管理。（官方指定的统一的配置文件）替换 requirements.txt 文件

## pyproject.toml

### 一个例子(TongYi-Z-Image)

`````python
[build-system]
requires = ["setuptools>=61.0"]
build-backend = "setuptools.build_meta"

[project]
name = "zimage-native"
version = "0.1.0"
description = "Z-Image PyTorch Native Implementation"
readme = "README.md"
requires-python = ">=3.10"
dependencies = [
    "torch>=2.5.0",
    "transformers>=4.51.0",
    "safetensors",
    "loguru",
    "pillow",
    "accelerate",
    "huggingface_hub>=0.25.0"
]

[project.optional-dependencies]
dev = [
    "black",
    "isort",
    "ruff"
]

[tool.setuptools.packages.find]
where = ["src"]
`````

这条命令用来安装虚拟环境的包：
````sh
pip install .
````

这个命令 执行过程 两步：
- 将当前的项目打包
- 根据 dependencies 安装 依赖

这也就会导致 将项目文件也打包到 虚拟环境 的三方库目录，要避免这个问题，加上一个 -e 参数即可，这会在 三方库目录下加入一个链接，指向源码：

````sh
pip install -e .
````

这样也还是不太好，每次加入一个包 都需要 查清楚版本号并 手动写入配置文件，太不优雅了，不如 pip install 来的方便。因此有了 Poetry、uv、PDM 等 工具 对 pip 和 venv 进行了 高级的封装。这些第三方工具底层用的也还是 pip 和 venv，但是提供了封装。
